Implement K-Means clustering on sales_data_sample.csv dataset. Determine the number of clusters using the elbow method.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Step 3: Select numerical columns only (K-Means can’t handle text)
#X = df.select_dtypes(include=['int64', 'float64'])


features = ['QUANTITYORDERED', 'PRICEEACH', 'SALES', 'MSRP']
X = df[features]
```

```python
import seaborn as sns

sns.boxplot(X)
plt.show()
```

```python
def remove_outliers(a):
    Q1 = a.quantile(0.25)
    Q3 = a.quantile(0.75)
    IQR = Q3-Q1

    lb = Q1-1*IQR
    ub = Q3+1*IQR

    clean_df = a[~((a<lb)|(a>ub)).any(axis=1)]
    return clean_df

df_clean = remove_outliers(X)
```

```python
import seaborn as sns

sns.boxplot(df_clean)
plt.show()
```

```python
X.isnull().sum()
```

```python
# Step 4: Handle missing values if any
X = X.dropna()
```

```python
X.isnull().sum()
```

```python
# Step 5: Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_clean)
```

### ELbow method for best K

```python
from sklearn.cluster import KMeans
wcss = []   # within-cluster sum of squares
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_) #the sum of squared distances from each sample to its assigned cluster centroid (i.e., WCSS).

plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.show()
```

### Apply Knn with optimal K

```python
best_k = 4
kmeans = KMeans(n_clusters=best_k, random_state=42)
df_clean['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
#Display results
print(df_clean[['Cluster']].value_counts())
```
