Implement Gradient Descent Algorithm to find
the local minima of a function.
For example, find the local minima of
thefunction y=(x+3)² starting from the point
x=2.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Define function and its derivative

```python
# Function: y = (x + 3)^2
def f(x):
    return (x + 3)**2

# Derivative: dy/dx = 2(x + 3)
def df(x):
    return 2 * (x + 3)
```

### Gradient Descent Algorithm

```python
# Initialize parameters
x = 2              # starting point
learning_rate = 0.1
iterations = 50

# Store x values to visualize later
x_list = [x]

# Perform Gradient Descent
for i in range(iterations):
    grad = df(x)
    x = x - learning_rate * grad #The learning rate (α) controls how big your step downhill is.
    x_list.append(x)

print("Local minimum occurs at x =", round(x, 4))
print("Minimum value of y =", round(f(x), 4))


#Gradient Descent is used to minimize the cost or loss function, i.e., 
#to make predictions as accurate as possible by adjusting model parameters (weights)
```

### Visualize

```python
# Plot function and gradient path
x_vals = np.linspace(-6, 4, 100)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label='y = (x + 3)^2')
plt.scatter(x_list, [f(i) for i in x_list], color='red', label='Gradient Descent Steps')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Descent to find Local Minima')
plt.legend()
plt.show()
```
